# 对换钱的贝叶斯分析
+ date: 2017-04-04
+ tags: math, Bayers

引用至[ “这山望着那山高”的数学原理 ](http://0x01.me/%E2%80%9C%E8%BF%99%E5%B1%B1%E6%9C%9B%E7%9D%80%E9%82%A3%E5%B1%B1%E9%AB%98%E2%80%9D%E7%9A%84%E6%95%B0%E5%AD%A6%E5%8E%9F%E7%90%86/)：
> 现在邀请你和一位路人甲来做一个游戏：我拿出两个信封分别递给你们，并告诉你
> 们一个装着的钱是另一个的两倍（但不知道哪个多哪个少）。你们有一次互相交换
> 的机会，想交换吗？然后打开信封，看一下自己拿到的钱数（但不要让对方知道），现在还想交换吗？

假设钱少的那个钱数为$y$，满足先验分布$\rho(y)$，某人抽到了钱数$x$。对应的后验：

\begin{align}
p(y=x|x)&=p(x|y=x)p(y=x)/p(x)\\
&=\frac{\rho(x)}{2}\\
p(y=x/2|x)&=p(x|y=x/2)p(y=x/2)/p(x)\\
&=\frac{\rho(x/2)}{2}\\
\end{align}

于是有$x$发生概率以及后验分布：

\begin{align}
p(x)&=\rho(x)+\rho(x/2)\\
p(y=x|x)&=\frac{\rho(x)}{\rho(x)+\rho(x/2)}\\
p(y=x/2|x)&=\frac{\rho(x/2)}{\rho(x)+\rho(x/2)}
\end{align}

换钱后钱增量$\Delta x$，
$$\mathrm{E}[\Delta x]=\frac{x\rho(x)-x\rho(x/2)/2}{\rho(x)+\rho(x/2)}$$

## 无脑交换的收益期望值
由于世界财富值有限，我们可以假定$\rho$是有界的，亦即
> $\exists A$使得对于$\forall x>A$，满足$\rho(x)\equiv 0$

那么
$$\begin{align}\overline{\mathrm{E}[\Delta x]}&=\int_0^\infty p(x)\mathrm{E}[\Delta x]dx\\
&=\int_0^\infty \Big[x\rho(x)-x\rho(x/2)/2\Big]dx\\
&=\lim_{A\to\infty}\int_0^A \Big[x\rho(x)-x\rho(x/2)/2\Big]dx\\
&=\lim_{A\to\infty}\int_0^A \Big[x\rho(x)-x\rho(x/2)/2\Big]dx\\
&=\lim_{A\to\infty}\int_{A/2}^A x\rho(x)dx\\
&=0
\end{align}$$
所以无脑交换在合理假设下无收益。
