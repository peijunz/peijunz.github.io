# GPA Converter
+ date: 2016-10-01
+ tags: ustc

```python
from bisect import bisect
def perct2gp(per):
    '''Convert percentage to grade points using bisect'''
    p=[60, 61, 64, 65, 68, 72, 75, 78, 82, 85, 90, 95]
    gpa=[0, 1, 1.3, 1.5, 1.7, 2.0, 2.3, 2.7, 3.0, 3.3, 3.7, 4.0, 4.3]
    ind=bisect(p, per)
    return gpa[ind]
```

```python
for i in range(80, 101, 5):
    print(i, '\t', perct2gp(i))
```
