# 拼图游戏
+ date: 2015-03-06

## 可解的必要条件
由置换数的奇偶性决定。逆序对定义为冒泡排序时候的相邻交换次数。

游戏中只能操纵空格子和相邻某个格子的对换。空格子移动一步的时候，横向/纵向逆序$(h, v)$变更一。如果最后空格子移动回到了远处，回来的时候，横向/纵向走的位移都是零，因此横序和纵序$(h, v)$都是偶的。因此奇数次置换不可解

## 充分性
我们考虑三元置换的可行性。首先如果我们有$ABC$和空格在一个角落里，那么显然很容易将$ABC$做轮换
```
xxxx   xxxx   xxxx
xxxx   xxxx   xxxx
xxAB   xxBC   xxCA
xxCO   xxAO   xxBO
```

对于任意的三个$ABC$：
+ 先将$ABC$和空格$O$，经过某个操作$U$拼到上述角落里面
+ 实行三元轮换操作$L$
+ 进行第一步的逆操作$U^{-1}$

至此，显然我们完成了三元轮换。唯一要证明的就是第一步的可行性
