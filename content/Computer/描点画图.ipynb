### 等距描点画图

\begin{align}
\Delta l&=\sqrt{\Delta x^2+\Delta y^2}=C\\
\Rightarrow \Delta x&=\frac{\Delta l}{\sqrt{1+y'^2}}
\end{align}

### 等角描点画图

\begin{align}
\theta&=\arctan\frac{\Delta y}{\Delta x}\\
\delta\theta&=C\\
\Rightarrow \Delta x&=\delta\theta\Big/\frac{y''}{1+y'^2}
\end{align}
