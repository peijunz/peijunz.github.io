### 重新分区/home
为安装系统等做准备
#### 困难：

+ 要想调整分区必须卸载/home
+ 使用中无法卸载

#### 解决方案

1. 先在控制台(Ctrl+Alt+F1)中输入`sudo init 1`，再按照提示输入密码就进入了单用户模式。
1. 输入`umount /home`取消挂载`/home`
1. 运行yast命令打开文本模式的yast(使用Tab和上下左右移动光标，Enter确认)
1. 进入System—–Partitioner
1. 按enter选中`/home`分区
1. 选中resize，按照提示进行操作
1. 输入`poweroff`重启
