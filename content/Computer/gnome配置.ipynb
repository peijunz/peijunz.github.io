## 任务栏不合并kde程序的bug
使用 https://github.com/bil-elmoussaoui/StartupWMClassFixer

```py
import glob
path = '/usr/share/applications/'
l = glob.glob(path + 'org.kde.*.desktop')
for i in l:
    k = i.split('.')[-2]
    print(k, 'org.kde.{}'.format(k), k, sep=', ')
```

运行上述代码追加内容至`database.csv`，再运行`./fix`

## 使用zsh和konsole
由于gnome-terminal的[BUG#706927](https://bugzilla.gnome.org/show_bug.cgi?id=706927)的存在, 虽然可以修改g-t的shell为zsh，但是并无法让nautilus打开g-t的时候自动跳转到当前目录。[曲线救国的方法](https://askubuntu.com/a/932969)是让g-t使用bash跳转到正确的目录，再从在这个bash里面跳转到konsole及其对应的zsh。即将以下代码加入`.bashrc`:
```sh
if ps -o cmd= -p $(ps -o ppid= -p $$) | grep -q gnome; then
  nohup terminator &> /dev/null &
  sleep 0.1s
  exit
fi
```
它会自动检测到当前运行bash的是gnome-terminal情况下，打开konsole。

**代价**:gnome-terminal充当了牺牲品，给更好用的konsole当配角。
