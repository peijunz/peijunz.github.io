````python
import numpy as np
def stem(d):
    "A stem-and-leaf plot that fits in a tweet by @ivanov"
    l,t=np.sort(d),10
    O=range(l[0]-l[0]%t,l[-1]+11,t)
    I=np.searchsorted(l,O)
    for e,a,f in zip(I,I[1:],O):
        print('%3d|'%(f/t),*(l[e:a]-f),sep='')
````

After definition of stem, we can now run to get results

````python
data = np.random.poisson(355, 113)
stem(data)
````
