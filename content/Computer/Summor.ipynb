# Summor with level
+ date: 2017-03-28
+ tags: python, algorithm

Ref: http://ewind.us/2016/sum-to-curry/

```python
class SumLv:
    def __init__(self, depth):
        self.depth=depth
        self.lv=self.depth
        self.s=0
    def __call__(self, *args):
        '''
        Empty call to return value
        Non-empty call to add values
        '''
        if args:
            self.s+=sum(args)
            self.lv-=len(args)
            if self.lv==0:
                return self.ret()
            elif self.lv<0:
                self.ret()
                raise Warning('Too many arguments!')
        else:
            self.ret() # Empty call invoking ret()
        return self
    def ret(self):
        '''Return the sum and reset level & sum'''
        self.lv=self.depth
        self.finalsum, self.s=self.s, 0
        return self.finalsum
    def __repr__(self):
        return "level={}, sum={}".format(self.lv, self.s)
```

```python
sum4=SumLv(4)
```

```python
print(sum4(1, 2, 3, 4))
print(sum4(1)(2, 3, 4))
print(sum4(1)(2)(3, 4))
print(sum4(1)(2)(3)(4))
```

```python
try:
    sum4(1,2,3,4,5)
except:
    print(sum4.finalsum)
```
